```python
import pandas as pd
```

```python
cmptns = pd.read_csv('meta-kaggle/Competitions.csv')

evlgs = pd.read_csv('meta-kaggle/EvaluationAlgorithms.csv')
evlgs.sample(3)
```

```python
cmptns.NumPrizes.value_counts()
```

```python
cmptns = pd.merge(cmptns, evlgs, left_on='EvaluationAlgorithmId', right_on='Id', suffixes=('', '_eval'))
cmptns.sample(3).T
```
